# Sort pandas Dataframe or a series

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
#sort 'genre' series in alphabetically descending order
movies['genre'].sort_values(ascending = False)
```

```python
#sort movies dataframe by a series ('duration') in decresing length
movies.sort_values('duration',ascending = False)
```

```python
#sort dataframe by two columns (first by 'content rating' and then by 'duration') 
movies.sort_values(['content_rating','duration'], ascending =[True,False])
```

# Filter rows of a pandas dataframe by column value

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
##filer dataframe with single condition :
#filter dataframe which have duration of the movie greater than and equal to 200 and genre  should be adventure
movies[movies['duration'] >= 200 ]  #output a filter dataframe
```

```python
##filter dataframe with multiple condition :
#filter dataframe having duration greater than 200 and genre should be 'Adventure'
movies[(movies['duration'] >= 200) & (movies['genre'] == 'Adventure')]    # &  and operator
```

```python
#filter rows having genre 'Adventure','Drama','Action'
movies[(movies['genre'] == 'Adventure') | (movies['genre'] == 'Drama') | (movies['genre'] == 'Action') ] # |  or operator
movies[movies['genre'].isin(['Adventure','Drama','Action'])]   #both code return the same value
```

# Explore pandas series


```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
#returns the data type of each column
movies.dtypes
```

```python
movies['genre'].describe()
```

```python
# return the frequency of each class
movies['genre'].value_counts()
```

```python
# returns a value counts of those greater than 100
movies['genre'].value_counts()[movies['genre'].value_counts() > 100]
```

```python
#return the % of each class
movies['genre'].value_counts(normalize = True)
```

```python
# return the name of unique class
movies['genre'].unique()
```

```python
#retuen the no. of unique class
movies['genre'].nunique()
```

```python
# returns the tally of the two series to be given in the argument
pd.crosstab(movies['genre'],movies['content_rating'])
```

```python
%matplotlib inline
movies['genre'].value_counts().plot(kind = 'bar')
```
